#### Import module

```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

#### Import Dataset

```python
url = r"https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/titanic-train.csv"
titanic = pd.read_csv(url)
```

#### Information about dataset

```python
print('Name of the Dataset: Titinic')
print('Number of Feature:', titanic.shape[1])
print('Number of Instance:', titanic.shape[0])
print('-'*35)
print(titanic.info())
```

#### Impotent Feature for making decision tree

```python
feature = ["Pclass","Sex","SibSp","Parch","Fare","Survived"]
data = pd.DataFrame(titanic[feature])
data['Sex'].replace({'female':0, 'male':1},inplace=True)
data = data.dropna()
data.head(3)
```

#### Train and test split

```python
x = data.drop('Survived', axis=1)
y = data['Survived']
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.3,random_state=42)
print('Training sample', train_x.shape)
print('Testing sample', test_y.shape)
```

#### Decision Tree Classification 

```python
model = DecisionTreeClassifier(criterion='gini',max_depth = 3)
model.fit(train_x, train_y)
```

#### Model Save into pickle file

```python
filename = "titanic_decisiontree.sav"
pickle.dump(model, open(filename, "wb"))
print('Model save into pickle file')
```

#### Load model from pickle file

```python
filename = "titanic_decisiontree.sav"
load_model = pickle.load(open(filename, "rb"))
print('Model load Succesfully !!')
```

```python
pred_y = load_model.predict(test_x)
df = pd.DataFrame({"Actual" : test_y, "Predicted" : pred_y})
df.head()
```

#### Clacification Report

```python
print(classification_report(pred_y, test_y, target_names=['Died','Survived']))
```
